```python
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
import numpy as np
import argparse
import cv2
import os
import keras
import imutils
import lime
from lime import lime_image
from skimage.segmentation import mark_boundaries
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
```

```python
model = load_model('covid19_model3.h5' )
print("Model Loaded Succesfully")
print(model.summary())
```

```python
image = cv2.imread("covid3.jpg")
plt.imshow(image)
```

```python
orig = cv2.resize(image, (224, 224))
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.resize(image, ( 224, 224))
#cv2.imshow("image", image)
image = np.array(image)/255.0
image = np.expand_dims(image, axis=0)
```

```python
pred = model.predict(image)
print(pred)
i = np.argmax(pred[0])
```

```python
image[-1].shape
```

```python
explainer = lime_image.LimeImageExplainer()
explanation = explainer.explain_instance(image[-1], model.predict,
                                             hide_color=0, num_samples=42)
```

```python
temp, mask = explanation.get_image_and_mask(
explanation.top_labels[1], positive_only=False, num_features=10, hide_rest=False)
#min_weight = 0.1)
plt.rcParams["figure.figsize"] = (20,10)
plt.figure()
plt.imshow(mark_boundaries(temp / 2 + 0.5, mask))
plt.show()
plt.savefig("LimeVisualizer.png")
```
